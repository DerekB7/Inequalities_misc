# Interpretation in terms of rare events and Boole's inequality


There is an interesting way of interpretting Bernouli's Inequality in terms of Boole's Inequaity / Union Bound.  This result is closely related to Exercise 1.47 in Gallagher's Stochastic Processes.  

Suppose we have iid Bernouli random variables -- in particular ones with with a very small probability of success are of interest here.  E.g. in Gallagher it's the case of where $p := 10^{-10}$.  


The probability of zero successes in (a large) number of Bernouli Trials is given by 

$(1-p)^n$

Consider defininig $A_i$ as the event of a success (aka 'heads') on the $ith$ coin toss.  

Then we are interesting in calculating the complement of the below:  

$Pr\{A_1 \bigcup A_2\bigcup... \bigcup A_n\}$ 

**applying Boole's Inequality** (aka the union bound), we have 

$Pr\{A_1 \bigcup A_2\bigcup... \bigcup A_n\} \leq Pr\{A_1\} +  Pr\{A_2\} + ...  Pr\{A_n\}$   

and the complement is given by negating and adding one, giving us 

$1 - n \cdot p  =1 - \Big(Pr\{A_1\} +  Pr\{A_2\} + ...  Pr\{A_n\}\Big) \leq  1 - Pr\{A_1 \bigcup A_2\bigcup...\bigcup A_n\} =  Pr\{A_1^C \cap A_2^C \cap ... \cap A_n^C \} = (1-p)^n$  

**And we see this is precisely the Bernouli Inequality for** $x \in[-1,0]$  

note that the mean number of sucesses is equal to 

$\lambda:= n\cdot p $ 

so we know   

$1 - \lambda \leq Pr\{A_1^C \cap A_2^C \cap ... \cap A_n^C \} = (1-\frac{n\cdot p}{n})^n = (1-\frac{\lambda}{n})^n \leq e^{-\lambda} = \text{Poisson probability estimate of no successes} $

$1 + b \leq \Big(1 + \frac{b}{n}\Big)^{n}$

Final leg:  
(1) recalling that $a_n = b + \mu$, notice that we have equality if $a_n = \mu$ (i.e. $b = 0$)    

(2) suppose $a_n < \mu$.  Because $a_n \geq 0$ this means $b \geq -\mu$ i.e. $b \geq -1$.  Alternatively we may have $a_n > \mu$ which means that $b > 0$.  Our inequality can thus be re-written as:

define $\frac{b}{n} := x$ and we rewrite this as:   
$1 + nx \leq \Big(1 + x\Big)^{n}$

we verify that real valued $x \geq -1$ (infact, per the above we know $x\geq \frac{-1}{n}$) in all cases (and natural number $n$), and hence we apply the Bernouli equality, which was proved in Exercise 2.2.

**extension:**  

As noted in the problem set, Akerberg's Refinement can also be a way of deriving $\text{GM} \leq \text{AM}$.  Specifically use it iteratively: 

recalling the general theorem is:
$a_n\Big(\frac{a_1 + a_2 + ... +a_{n-1}}{n-1}\Big)^{n-1} \leq  \Big(\frac{a_1 + a_2 + ... +a_{n-1} +a_{n}}{n}\Big)^{n}$


we start small:
$a_2\Big(\frac{a_1}{1}\Big)^{1} \leq  \Big(\frac{a_1 + a_2}{2}\Big)^{2}$  
equivalently

$a_2 a_1 \leq  \Big(\frac{a_1 + a_2}{2}\Big)^{2}$ 
(and square root to see  $\text{GM} \leq \text{AM}$ for 2 variables.) 

now increment to $n=3$  
$ a_3\Big(\frac{a_1 + a_2}{2}\Big)^{2} \leq  \Big(\frac{a_1 + a_2 + a_{3}}{3}\Big)^{3}$
which we can combine with the earlier result and see:  

$a_3 \big(a_2a_1\big) \leq a_3\Big(\frac{a_1 + a_2}{2}\Big)^{2} \leq  \Big(\frac{a_1 + a_2 + a_{3}}{3}\Big)^{3}$   

and now cube root to see $\text{GM} \leq \text{AM}$.

and for the case of 4 we have

$a_4\Big(\frac{a_1 + a_2 +a_{3}}{3}\Big)^{3} \leq  \Big(\frac{a_1 + a_2 +a_{3} +a_{4}}{4}\Big)^{4}$

then apply what we know above:

$a_4 \big(a_3 a_2a_1\big) \leq a_4\Big(\frac{a_1 + a_2 +a_{3}}{3}\Big)^{3} \leq  \Big(\frac{a_1 + a_2 +a_{3} +a_{4}}{4}\Big)^{4}$

and so on
