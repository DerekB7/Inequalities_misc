# Interpretation in terms of rare events and Boole's inequality


There is an interesting way of interpretting Bernouli's Inequality in terms of Boole's Inequaity / Union Bound.  This result is closely related to Exercise 1.47 in Gallagher's Stochastic Processes.  

Suppose we have iid Bernouli random variables -- in particular ones with with a very small probability of success are of interest here.  E.g. in Gallagher it's the case of where $p := 10^{-10}$.  


The probability of zero successes in (a large) number of Bernouli Trials is given by 

$(1-p)^n$

Consider defining $A_i$ as the event of a success (aka 'heads') on the $ith$ coin toss.  

Then we are interesting in calculating the complement of the below:  

$Pr\{A_1 \bigcup A_2\bigcup... \bigcup A_n\}$ 

**applying Boole's Inequality** (aka the union bound), we have 

$Pr\{A_1 \bigcup A_2\bigcup... \bigcup A_n\} \leq Pr\{A_1\} +  Pr\{A_2\} + ...  Pr\{A_n\}$   

and the complement is given by negating and adding one, giving us 

$1 - n \cdot p  =1 - \Big(Pr\{A_1\} +  Pr\{A_2\} + ...  Pr\{A_n\}\Big) \leq  1 - Pr\{A_1 \bigcup A_2\bigcup...\bigcup A_n\} =  Pr\{A_1^C \cap A_2^C \cap ... \cap A_n^C \} = (1-p)^n$  

- - - - 
another way at this result, which is perhaps more pleasant, is to consider $n$ **independent** coin tosses, each with probability $p_i$ of heads and the event that the $ith$ toss is heads is given by $A_i$ and the probability that amongs the n heads there is at least one heads is given by 

$A := A_1 \bigcup A_2 \bigcup ... \bigcup A_n$   
and of course 
$A^c$ is the event that none of the tosses results in heads 

**And we see this is precisely the Bernouli Inequality for** $x \in[-1,0]$  

so we have 

$1 = P\big(A \bigcup A^c\big) = P\big(A\big) + P\big(A^c\big) =  P\big(A_1 \bigcup A_2 \bigcup ... \bigcup A_n\big) + P\big(A^c\big)$  
$\leq \Big(p_1 + p_2 + ... + p_n\Big) + P\big(A^c\big)= \Big(p_1 + p_2 + ... + p_n\Big) + \big(1-p_1\big)\big(1-p_2\big)...\big(1-p_n\big)$    
(by union bound / Boole's Inequality)  

re-arranging terms we get 

$1 - \Big(p_1 + p_2 + ... + p_n\Big) \leq \big(1-p_1\big)\big(1-p_2\big)...\big(1-p_n\big)$   

which is the inhomogenous case of the Bernouli Inequality.  When the amounts are homogenous we recover the Bernouli Inequality for $x = x_i = -pi \in [-1,0]$  


note that the mean number of sucesses is equal to 

$\lambda:= n\cdot p $ 

so we know   

$1 - \lambda \leq Pr\{A_1^C \cap A_2^C \cap ... \cap A_n^C \} = (1-\frac{n\cdot p}{n})^n = (1-\frac{\lambda}{n})^n \leq e^{-\lambda} = \text{Poisson probability estimate of no successes} $  

** further remark using insights from chapters 12 and 13** 
The inhomogenous case can be interpretted as a *refinement* of the Bernouli inequality.  Because elementary symmetric functions (of non-negative numbers) are Schur concave, and where 

$\bar{p} := \frac{p_1 + p_2 +.... + p_n}{n}$   

we have 

$1 - n\cdot \bar{p} $  
$= 1 - \Big(p_1 + p_2 + ... + p_n\Big) $  
$\leq \big(1-p_1\big)\big(1-p_2\big)...\big(1-p_n\big) $  
$\leq (1-\bar{p}\big)^n$   
where the first inequality follows by Boole's Inequality and the second could be said to follow from Schur concavity.  However in this case, we can *also* simply justify the result by $\text{GM} \leq \text{AM}$ which tells us 

$\Big(\big(1-p_1\big)\big(1-p_2\big)...\big(1-p_n\big)\Big)^\frac{1}{n} \leq \frac{1}{n}\sum_{i=1}^n (1-p_i) = \big(1-\bar{p}\big)$  
hence raising each side to the nth power gives the upper bound.  


**much much later remark**  

yet another way of getting the inhomogenous bernouli inequality lower bound that we derived above via union bound, is naturally to use the theory of majorization  

that is, we revisit:  

$1 - \Big(p_1 + p_2 + ... + p_n\Big) \leq \big(1-p_1\big)\big(1-p_2\big)...\big(1-p_n\big)$    
for $p_i \in (0,1)$.  

To streamline the argument, *we assume WLOG that* $p_1 \geq p_2 \geq ... \geq p_n$   

note that if $p_1 + p_2 + ... + p_n \geq 1$ the bound automatically holds, because it merely says that a positive number is bounded below by a non-positive number. So it suffices to prove the case where $p_1 + p_2 + ... + p_n \in (0,1)$  

using the concavity (and hence Schur concavity) of $\log\big(1-x_k\big)$ for $x_k \in [0,1)$  
see e.g. second derivative test here:  https://www.wolframalpha.com/input/?i=2nd+derivative+ln(1-x)  

and the majorization relationship given here  

$\mathbf a  =\begin{bmatrix}p_1 + p_2 + ... + p_n\\ 
0\\ 
\vdots\\ 
0\\  
\end{bmatrix} \succ \begin{bmatrix}p_1 \\ 
p_2\\ 
\vdots\\ 
p_n\\  
\end{bmatrix}= \mathbf b$   


and consider the symmetric mapping given by $f: \mathbb R^n \longrightarrow \mathbb R $  
(where the domain is technically given component-wise for values in $[0,1)$  

$f\big(\mathbf x\big) = \sum_{k=1}^n \log\big(1-x_k\big)$  
so by Schur concavity 

$\log\big(1-(p_1 + p_2 + ... + p_n)\big) $  
$= \log\big(1-(p_1 + p_2 + ... + p_n)\big) + 0 + ... + 0  $  
$= \log\big(1-(p_1 + p_2 + ... + p_n)\big) + \log\big(1-0\big) + ... + \log\big(1-0\big)  $  
$= f\big(\mathbf a\big) $  
$\leq f\big(\mathbf b\big) $  
$ = \log\big(1-p_1\big) + \log\big(1-p_2\big)  + ... + \log\big(1-p_n\big)$  
$ = \log\Big(\big(1-p_1\big)\big(1-p_2\big)... \big(1-p_n\big)\Big)$    

exponentiating both sides then gives  

$1-\big(p_1 + p_2 + ... + p_n\big) \leq \big(1-p_1\big)\big(1-p_2\big)... \big(1-p_n\big)$   

as desired.  

*remark:*  
The Boole's Inequality / Union Bound approach is perhaps more elementary and more easily to generalize (via Bonferonni's Inequalities).  However, it is satisfying to derive the same result multiple ways and as is often the case, majorization nicely allows us to do exactly this.  



**additional remark:**   
from a foundational standpoint in Analysis, we may find the Bernouli Inequality to be quite useful.  E.g. suppose we want to prove that the geometric series converges absolutely for $x \in (0,1)$ (which we can use to bound the tail on exponential function's power series which proves it converveges absolutely, and then we can go build out much more of analysis).

suppose we have $z := \frac{y}{x}$


then the finite geometric series of $z$ is given by:


$f(z)  = (z^0 + z + z^2 + z^3 + .... + z^{n})$    

notice that  
$x^n f(z) = x^{n}y^0 + x^{n-1}y^{1} + x^{n-2}y^{2} + ... +x^{2}y^{n-2} + x^{1}y^{n-1} + x^{0}y^{n}$    

$(1-z)f(z)  = (1-z)(1 + z + z^2 + z^3 + .... + z^{n})= 1 - z^{n+1}$

$f(z) = \frac{1 - z^{n+1}}{1-z}$

$f(z) = \frac{1 - \frac{y^{n+1}}{x^{n+1}}}{1-\frac{y}{x}} = \frac{\frac{x^{n+1}}{x^{n+1}} - \frac{y^{n+1}}{x^{n+1}}}{\frac{x}{x}-\frac{y}{x}} = \frac{x}{x^{n+1}}\big(\frac{x^{n+1} - y^{n+1}}{x - y}\big) = \frac{1}{x^{n}}\big(\frac{x^{n+1} - y^{n+1}}{x - y}\big)$

hence we see:

$ x^n f(z) = x^n \frac{1}{x^{n}}\big(\frac{x^{n+1} - y^{n+1}}{x - y}\big) =\frac{x^{n+1} - y^{n+1}}{x - y} = x^{n}y^0 + x^{n-1}y^{1} + x^{n-2}y^{2} + ... +x^{2}y^{n-2} + x^{1}y^{n-1} + x^{0}y^{n}$

note: the series in the back was *slightly* different than the above... it is: 

$\frac{x^{n} - y^{n}}{x - y} = x^{n-1}y^0 + x^{n-2}y^{1} + x^{n-3}y^{2} + ... +x^{2}y^{n-3} + x^{1}y^{n-2} + x^{0}y^{n-1}$

# Ex 2.11

(This is the official solution)


for non-negative $a_k$ and $b_k$ and natural numbers $k$ and $n$ where $1 \leq k\leq n$

**claim:**  

$\Big(\prod_{k=1}^n a_k\Big)^{\frac{1}{n}}+ \Big(\prod_{k=1}^n b_k\Big)^{\frac{1}{n}} \leq \Big(\prod_{k=1}^n \big(a_k +b_k\big)\Big)^{\frac{1}{n}}$

This reads as

$\text{GM}\Big(\mathbf a\Big) + \text{GM}\Big(\mathbf b\Big) \leq \text{GM}\Big(\big(\mathbf a + \mathbf b\big)\Big)$

**proof:**

First notice that if all $a_k = 0$ and all $b_k=0$ then it is an equality at zero.  We now assume at least one non-zero $a_k$ or $b_k$ for each k, such that the Right Hand Side is $\gt 0$, and thus we may divide by the RHS without changing the inequality.


$\frac{\Big(\prod_{k=1}^n a_k\Big)^{\frac{1}{n}}+ \Big(\prod_{k=1}^n b_k\Big)^{\frac{1}{n}}}{\Big(\prod_{k=1}^n \big(a_k +b_k\big)\Big)^{\frac{1}{n}}} \leq 1$

now re-write this as:

$\Big(\frac{\prod_{k=1}^n a_k}{\prod_{k=1}^n a_k +b_k}\Big)^{\frac{1}{n}} + \Big(\frac{\prod_{k=1}^n b_k}{\prod_{k=1}^n a_k +b_k}\Big)^{\frac{1}{n}}\leq 1$

$\Big(\prod_{k=1}^n \big(\frac{a_k}{a_k +b_k}\big) \Big)^{\frac{1}{n}} + \Big(\prod_{k=1}^n \big(\frac{b_k}{a_k +b_k}\big) \Big)^{\frac{1}{n}}\leq 1$


now we use $\text{GM} \leq \text{AM}$


$\Big(\prod_{k=1}^n \big(\frac{a_k}{a_k +b_k}\big) \Big)^{\frac{1}{n}} + \Big(\prod_{k=1}^n \big(\frac{b_k}{a_k +b_k}\big) \Big)^{\frac{1}{n}}\leq  \Big(\frac{1}{n}\sum_{k=1}^n \frac{a_k}{a_k+b_k}\Big) + \Big(\frac{1}{n} \sum_{k=1}^n \frac{b_k}{a_k+b_k}\Big) = \sum_{k=1}^n \frac{1}{n}\Big(\frac{a_k +b_k}{a_k+b_k}\Big)= \sum_{k=1}^n \frac{1}{n} = 1$


**commentary:**

This is a shockingly effective use of $\text{GM} \leq \text{AM}$ and a very simple solution to what seems to be a quite powerful result. 


# Exercise 2.10 (Akerberg's Refinement)

for real non-negative numbers $a_1, a_2, ..., a_n$ and natural number $n\geq 2$, the following bound exists:

**claim:**  
$a_n\Big(\frac{a_1 + a_2 + ... +a_{n-1}}{n-1}\Big)^{n-1} \leq  \Big(\frac{a_1 + a_2 + ... +a_{n-1} +a_{n}}{n}\Big)^{n}$ 

**proof:**
First notice that  all $a_i \geq 0$.  In the special case that all $a_i =0$  the inequality becomes an equality.  In the event all $a_i=0$ except $a_n > 0$ then there is a strict inequality.  Thus the claim is proved if $a_i =0$ for $i=\{1,2, ..., n-1\}$. The rest of this posting assume at least one of those values is $>0$. 

Now define: 

$\mu := \frac{a_1 + a_2 + ... +a_{n-1}}{n-1}$

For all remaining cases to consider we know $\mu >0$ and hence there is some postive scalar $\gamma^{n}$ we may multiply each side by (which will not alter the inequality) such that we have 

$\mu^* := \gamma \frac{a_1 + a_2 + ... +a_{n-1}}{n-1} = 1$  
and we would also have $a_n^* = \gamma a_n$

after proving the claim with $\gamma$ in place (i.e. operating on $\mu*$ and $a_n*$) we may divide it out at the end which would prove that the claim holds for any $\mu \gt 0$.  **Thus without a loss of generality we may assume ** $\mu = 1$. 

$b: = a_n - \mu =  a_n - 1$, or equivalently  
$a_n = b + \mu = b +1$

$LHS = (b +\mu)\mu^{n-1} = \mu^{n-1}b +\mu^n = b + 1$

$RHS = \Big(\frac{a_1 + a_2 + ... +a_{n-1} +(\mu + b)}{n}\Big)^{n} = \Big(\frac{a_1 + a_2 + ... +a_{n-1} +\mu}{n} + \frac{b}{n}\Big)^{n} = \Big(\mu + \frac{b}{n}\Big)^{n} = \Big(1 + \frac{b}{n}\Big)^{n}$


Thus we need to prove:

$1 + b \leq \Big(1 + \frac{b}{n}\Big)^{n}$

Final leg:  
(1) recalling that $a_n = b + \mu$, notice that we have equality if $a_n = \mu$ (i.e. $b = 0$)    

(2) suppose $a_n < \mu$.  Because $a_n \geq 0$ this means $b \geq -\mu$ i.e. $b \geq -1$.  Alternatively we may have $a_n > \mu$ which means that $b > 0$.  Our inequality can thus be re-written as:

define $\frac{b}{n} := x$ and we rewrite this as:   
$1 + nx \leq \Big(1 + x\Big)^{n}$

we verify that real valued $x \geq -1$ (infact, per the above we know $x\geq \frac{-1}{n}$) in all cases (and natural number $n$), and hence we apply the Bernouli equality, which was proved in Exercise 2.2.

**extension:**  

As noted in the problem set, Akerberg's Refinement can also be a way of deriving $\text{GM} \leq \text{AM}$.  Specifically use it iteratively: 

recalling the general theorem is:
$a_n\Big(\frac{a_1 + a_2 + ... +a_{n-1}}{n-1}\Big)^{n-1} \leq  \Big(\frac{a_1 + a_2 + ... +a_{n-1} +a_{n}}{n}\Big)^{n}$


we start small:
$a_2\Big(\frac{a_1}{1}\Big)^{1} \leq  \Big(\frac{a_1 + a_2}{2}\Big)^{2}$  
equivalently

$a_2 a_1 \leq  \Big(\frac{a_1 + a_2}{2}\Big)^{2}$ 
(and square root to see  $\text{GM} \leq \text{AM}$ for 2 variables.) 

now increment to $n=3$  
$ a_3\Big(\frac{a_1 + a_2}{2}\Big)^{2} \leq  \Big(\frac{a_1 + a_2 + a_{3}}{3}\Big)^{3}$
which we can combine with the earlier result and see:  

$a_3 \big(a_2a_1\big) \leq a_3\Big(\frac{a_1 + a_2}{2}\Big)^{2} \leq  \Big(\frac{a_1 + a_2 + a_{3}}{3}\Big)^{3}$   

and now cube root to see $\text{GM} \leq \text{AM}$.

and for the case of 4 we have

$a_4\Big(\frac{a_1 + a_2 +a_{3}}{3}\Big)^{3} \leq  \Big(\frac{a_1 + a_2 +a_{3} +a_{4}}{4}\Big)^{4}$

then apply what we know above:

$a_4 \big(a_3 a_2a_1\big) \leq a_4\Big(\frac{a_1 + a_2 +a_{3}}{3}\Big)^{3} \leq  \Big(\frac{a_1 + a_2 +a_{3} +a_{4}}{4}\Big)^{4}$

and so on
